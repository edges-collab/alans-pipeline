# Compare outputs from the H2Case run with the "correct" outputs directly from Alan

```python
import numpy as np
import matplotlib.pyplot as plt
from edges_cal.alanmode import read_specal
```

```python
orig = read_specal("results/specal_final_case2.txt")
```

```python
testcase = read_specal("scripts/H2Case/specal.txt")
```

```python
testcase = {name: np.round(testcase[name], decimals=6) for name in testcase.dtype.names}  # six decimals in the orig
```

```python
for name in testcase:
    print(name, np.sum(np.abs(orig[name] - testcase[name])))
```
